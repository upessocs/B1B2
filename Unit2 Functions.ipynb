```python
def sum3(a, *args): # *args *values
    """
    this function takes one mandatory argument 
    and other optional arguments

    and return its sum
    
    """
    print(a)
    print(args)
    res=a
    for i in range(0,len(args)):
        res = res + args[i]
    
    return res
```
